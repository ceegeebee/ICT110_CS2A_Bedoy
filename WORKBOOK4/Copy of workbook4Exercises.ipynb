```python
%%writefile ownmodule.py

"""
My Own Module
"""

# Define some variables
favorite_number = 30
greeting = "Konnichiwa"


# Define a function to double a number
def double_number(x):
    """
    Doubles the input number.

    Args:
        x (int or float): The number to be doubled.

    Returns:
        int or float: The result of doubling the input number.
    """
    return x * 2

# Define a function to greet a person
def greet_person(name):
    """
    Generates a personalized greeting message for the given name.

    Args:
        name (str): The name of the person to greet.

    Returns:
        str: The greeting message.
    """
    return f"Hello, {name}! Nice to meet you."
```
