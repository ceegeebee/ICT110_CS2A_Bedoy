### Exercise

A. Create an ndarry object (A) with any values from 1 to 100 in a nested Python List. You can choose your range from the given values (1-100).

* Given that A is an ndarray object, many functionalities in numpy could be accessed by:
    + Make sure to explain each line of codes that you have created.

1. Using a property of the ndarray object
      - e.g. `A.shape`
2. Using a method of the ndarray object
      - e.g. `A.max()`
3. Using a numpy function and passing an ndarray
     - e.g. `np.max(A)`

B. Create a 5x5 integer-array filled with ones.
* Check its shape, size, and number of dimensions. Explain briefly the codes you have created by writing a comment beside the code.

C. Create an array of values from -5 to + 5. Multiply each value with 2. Print the resulting array.

```python
import numpy as np
# A. Create an ndarray object (A) with values from 1 to 100 in a nested Python List.
A = np.array([[i for i in range(1, 101)]])  # Creates a 1x100 ndarray with values from 1 to 100

# Property of the ndarray object: Shape of the array
print("\nShape of A:", A.shape)  # Prints the shape of A

# Method of the ndarray object: Maximum value in the array
print("\nMaximum value in A:", A.max())  # Prints the maximum value in A

# Numpy function: Maximum value in the array
print("\nMaximum value in A (using np.max):", np.max(A))  # Prints the maximum value in A using np.max()

# B. Create a 5x5 integer-array filled with ones.
ones_array = np.ones((5, 5), dtype=int)  # a 5x5 array filled with ones


print("\nShape of ones_array:", ones_array.shape)  # Prints the shape of ones_array
print("\nSize of ones_array:", ones_array.size)    # Prints the size of ones_array
print("\nNumber of dimensions of ones_array:", ones_array.ndim)  # Prints the number of dimensions of ones_array

# C. Create an array of values from -5 to +5. Multiply each value by 2.
values = np.arange(-5, 6)  # Creates an array of values from -5 to +5
result_array = values * 2  # Multiplies each value in the array by 2

# Prints the resulting array
print("\nResulting array after multiplying each value by 2:", result_array)
```

```python
#In this result four-dimensional array of random integer is displayed chosen from the interval [0, 12]
four_dimAr = np.random.randint(0, 13, size=(3, 4, 5, 6))
print("Four-dimensional array:")
print(four_dimAr )


#The output for "Array Dimensions:" displays the number of dimensions of the array, which is 4.

print("\nArray Dimensions:")
print(four_dimAr.ndim)

#The output for "Array Shape:" displays the shape of the array. The shape is (3, 4, 5, 6), which means the array has:
# 3 elements along the first axis,
# 4 elements along the second axis,
# 5 elements along the third axis,
# 6 elements along the fourth axis.
print("\nArray Shape:")
print(four_dimAr .shape)

# The output for "Array Size:" displays the total number of elements in the array, which is 360 (3 * 4 * 5 * 6).
print("\nArray Size:")
print( four_dimAr .size)

#The result displays maximum number for each dimensions.
print("\nMaximum number for each dimension:")
max_perDim = [np.max(four_dimAr [:, :, :, i]) for i in range(four_dimAr .shape[3])]
print(max_perDim )


# the result dispalys the multiplied four-dimensional array by 4 and stores the result in a new array called multiplied_array.
multiplied_array = four_dimAr  * 4
print("\n\nMultiplied array by 4:")
print(multiplied_array)
```
